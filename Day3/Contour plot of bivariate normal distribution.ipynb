```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#initialize parameters
M_u = 0
M_v = 0
Sig_u = 1.0
Sig_v = 1.0
rho = 0
```

```python
#Define a function over x and y
def f(u,v):
    return((1/np.sqrt(2*np.pi))*(1/Sig_u)*np.exp(-0.5*(u-M_u)**2/Sig_u**2*(1-rho**2))*
    (1/np.sqrt(2*np.pi))*(1/Sig_v)*np.exp(-0.5*(v-M_v)**2/Sig_v**2*(1-rho**2))*
    (1/np.sqrt(1-rho**2))*np.exp(-rho*(u-M_u)*(v-M_v)/(Sig_u*Sig_v*(1-rho**2))))
    
```

```python
u=np.linspace(-3,3,100)
v=np.linspace(-3,3,100)
```

```python
X,Y = np.meshgrid(u,v)
```

```python
plt.contourf(X,Y,f(X,Y),20,cmap='magma')
plt.colorbar()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,f(X,Y),20,cmap='magma')
plt.show()
```
