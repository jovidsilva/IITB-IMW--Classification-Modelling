```python
import pandas as pd
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
```

```python
p_val=0.8
N=100
X=sp.bernoulli.rvs(size=(N,1),p=p_val)
```

```python
print(X.flatten())
```

```python
plot=plt.pie(X.flatten())
```

```python
df = pd.DataFrame(X)
df.columns = ['Values']
print(df.value_counts())
```

```python
df2 = pd.DataFrame(df.value_counts())
df2 = df2.reset_index()
df2.columns=['Values','Frequency']
df2.iloc[:,:]
```

```python
plot2=plt.pie(df2['Frequency'])
```

```python
plot3=df2.plot.bar(x='Values',y='Frequency')
plot3
```
